### Import the Data

Let's extract the our **Titanic data** from the .xls file, create a  pandas DataFrame and look at the available indicators:


- ***pclass***: Socio-economic class (1 = Upper class; 2 = Middle class; 3 = Lower class)
- ***survived***: Outcome of survival (0 = No; 1 = Yes)
- ***name***: Name of passenger
- ***sex***: Sex of the passenger
- ***age***: Age of the passenger (Some entries contain NaN)
- ***sibsp***: Number of siblings and spouses of the passenger aboard
- ***parch***: Number of parents and children of the passenger aboard
- ***ticket***: Ticket number of the passenger
- ***fare***: Fare paid by the passenger
- ***cabin***: Cabin number of the passenger (Some entries contain NaN)
- ***embarked***: Port of embarkation of the passenger (C = Cherbourg; Q = Queenstown; S = Southampton)
- ***boat***: Lifeboat  (if survived)
- ***body***: Body number (if did not survive and body was recovered)
- ***home.dest***: Home/destination

```python
url_excel = "https://github.com/ubbamiller/brownbag/blob/main/titanic3.xls?raw=true"  #titantic dataset put on github

titantic_df = pd.read_excel(url_excel) #read the datafile into a dataframe. Could be any kind of file

titantic_df_copy = titantic_df.copy()

titantic_df.head(5) #look at first 5 records
```

```python
#load a CSV file using panda
titantic_df.info() # data types
titantic_df.describe() # statistical information
titantic_df.isnull().sum()
```

## 7. <u>Data wrangling </u>
***Data wrangling is the process of cleaning and unifying messy and complex data sets for easy access and analysis***

### 7.a. <u>Data Mapping</u>
**Note that the passenger class(pclass) and survived are not intuitive. Sometimes want to replace these values with something meaningful.**

```python
titantic_df['survived_map'] = titantic_df['survived'].map({0:'no', 1:'yes'}) 
titantic_df['pclass_map'] = titantic_df['pclass'].map({1:'upper', 2:'middle', 3:'lower'}) 
titantic_df['embarked'] = titantic_df['embarked'].map({'S':'Southhampton', 'C':'Cherbourg', 'Q':'Queenstown'}) 
titantic_df.head(5)
titantic_df_copy.head(5)
```

### 7.b. Finding blanks and replacing them with Nan

***If you have values in your dataframe that are blank, you want to be able to find them and exclude them from analysis. Best way(IMHO) is to replace them with NaN***

https://stackoverflow.com/questions/13445241/replacing-blank-values-white-space-with-nan-in-pandas

```python
import numpy as np
titantic_df = titantic_df.replace(r'^\s*$', np.nan, regex=True)
titantic_df.isnull().sum()
```

## 8. Methods for querying data in pandas



```python
print("DF using query method")
titantic_df.query('survived_map =="yes" & sex=="female"').head(5)

print("DF using dataframe true and false")
titantic_df[(titantic_df['survived_map']=='yes') & (titantic_df['sex']== 'female')].head(5)

print("DF using loc to control column output")
titantic_df.loc[(titantic_df['survived_map']=='yes') & (titantic_df['sex']== 'female'),["name", "sex","age"]].head(5)


```

### 12. More dataframe querying - two tables using sqldf

https://www.analyticsvidhya.com/blog/2021/07/pandasql-best-way-to-run-sql-queries-and-codes-in-jupyter-notebook-using-python/

```python
import pandas as pd
################# set pandas display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
##############################################

######pandasql libraries to import############
from pandasql import sqldf
from pandasql import load_meat, load_births
#############################################

######Load data)
meat = load_meat()
births = load_births()

#meat.head(5)
#births.head(5)

### helper function for running sql########
pysqldf = lambda q: sqldf(q,globals())
###########################################

###Running some sql as examples##########################
print('first 5 rows of meat')
result = pysqldf("Select * from meat LIMIT 5;")
result.head()
print('first 5 birth rows')
result2 = pysqldf("Select * from births LIMIT 5;")
result2.head()

print("inner join on beef and births")
# inner join
q = """SELECT
        m.date, m.beef, b.births
     FROM
        meat m
     INNER JOIN
        births b
           ON m.date = b.date;"""
print(pysqldf(q).head(5))

#group by example
query = '''SELECT
            strftime('%Y', date) as year
           , SUM(beef) as beef_total
           FROM
              meat
           GROUP BY
              year
              LIMIT 5;
    '''

print("\nBeef by year total")
print(pysqldf(query).head(5))
```

### 13.finding duplicate rows in a single table

### <u>14.Comparing two dataframes using datacompy library</u>

https://www.geeksforgeeks.org/compare-pandas-dataframes-using-datacompy/

https://capitalone.github.io/datacompy/api/datacompy.html

It takes **two dataframes** as input and gives us a **human-readable report** containing statistics that lets us know the **similarities and dissimilarities** between the two dataframes.

```python
# pip install datacompy # library for comparing two dataframes
#!pip install datacompy
from io import StringIO
import datacompy  

data1 = """employee_id, name, job
1, rajiv kapoor, tester
2, rahul agarwal,code
3, alice johnson,hr
4, richard ashwell,developer
"""
   
data2 = """employee_id, name
1, rajiv khanna
2, rahul aggarwal
3, alice tyson
5, phil hodgson
"""
   
df1 = pd.read_csv(StringIO(data1))
df2 = pd.read_csv(StringIO(data2))
   
compare = datacompy.Compare(
    df1, #first dataframe
    df2, #second dataframe
      
    # columns or list to join dataframe on
    join_columns = 'employee_id', 
      
    # absolute tolerance 
    # #if two values are different how much difference tolerated #
    # before reporting difference
    # defaults to 0 if not defined
    abs_tol = 0, #definition abs(x-y)
      
    # Optional, defaults to 0
    rel_tol = 0, #abs(X - Y) / min(abs(X), abs(Y))
      
    # Optional, defaults to 'df1'
    df1_name = 'Original',
      
    # Optional, defaults to 'df2'
    df2_name = 'New' ,
     #Flag to strip whitespace
    ignore_spaces=True,
    #Flag to ignore the case of string columns
    ignore_case=True,
    cast_column_names_lower=True,
    
    )
  
# if ignore_exra_columns=True, 
# the function won't return False
# in case of non-overlapping 
# column names
compare.matches(ignore_extra_columns = False)   
   
# This method prints out a human-readable 
# report summarizing and sampling 
# differences
print(compare.report())


```
